## Let's take the first target and drop the second.

```python
temperature = temp.copy()
```

```python
temperature
```

```python
temp1 = temperature.drop('Next_Tmin', 1)
```

```python
x = temp1.drop(['Next_Tmax'], 1)
y = temp1['Next_Tmax']
```

```python
x
```

```python
y
```

### Standard Scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x = sc.fit_transform(x)
```

```python
pd.DataFrame(x)
```

The data is brought to one similar scale.

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
def maxr2_score(mod,x,y):
    max_r_score=0
    for r in range(42,100):
        x_train, x_test, y_train, y_test = train_test_split(x, y,random_state = r,test_size=0.20)
        mod.fit(x_train,y_train)
        y_pred = mod.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print("r2 score corresponding to ",r," is ",r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r=r
    print("max r2 score corresponding to ",final_r," is ",max_r_score)
    return final_r
```

```python
#Cross_val_score fucntion 

from sklearn.model_selection import cross_val_score
def model_evaluation(model,x,y):
    c_scores=cross_val_score(model,x,y,cv=5,scoring="r2")
    print("Mean r2 score for regressor: ",c_scores.mean())
    print("standard deviation in r2 score for regressor: ",c_scores.std())
    print(c_scores)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression

le=LinearRegression()
r = maxr2_score(le,x,y)
```

```python
print("LinearRegression\n\n")
model_evaluation(le,x,y)
```

### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore")
rfr=RandomForestRegressor()
parameters = {"n_estimators":[10,100,500]}
clf = GridSearchCV(rfr, parameters, cv=5,scoring="r2")
clf.fit(x,y)
clf.best_params_
```

```python
#lets check Random forest using n_estimators=500

from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(n_estimators=500)
#Lets chcek the cross_val_score for Random Forest Regressor
print("Random Forest Regressor\n\n")
model_evaluation(rfr,x,y)
```

### KNN

```python
from sklearn.neighbors import KNeighborsRegressor

knr=KNeighborsRegressor()
parameters = {"n_neighbors":range(2,30)}
clf = GridSearchCV(knr, parameters, cv=5,scoring="r2")
clf.fit(x, y)
clf.best_params_
```

```python
knr=KNeighborsRegressor(n_neighbors=29)
maxr2_score(knr,x,y)
```

```python
#Lets check the cross_val score of KKN regressor
print("KNN Regressor\n\n")
model_evaluation(knr,x,y)
```

### SVM

```python
from sklearn.svm import SVR
svr=SVR()
parameters={"kernel":["linear", "poly", "rbf"],"C":[0.001,0.01,0.1,1,10]}
clf = GridSearchCV(svr, parameters, cv=5,scoring="r2")
clf.fit(x,y)
clf.best_params_
```

```python
svr=SVR(C=1,kernel="linear")
maxr2_score(svr,x,y)
```

Out of all the models, SVR does give us the max r2 score of 0.79.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state = 79,test_size=0.20)
svr.fit(x_train,y_train)
y_pred=svr.predict(x_test)
```

```python
print("r2_score is: ",r2_score(y_test,y_pred))
```

```python
Next_Max_Temp_prediction = svr.predict(x_test)
```

```python
#lets make the dataframe for Next_Max_Temp_prediction

Next_Max_Temp_prediction=pd.DataFrame(Next_Max_Temp_prediction,columns=["Next_Tmax"])
```

```python
#Lets save the submission to csv

Next_Max_Temp_prediction.to_csv("Next_Max_Temp_prediction.csv",index=False)
```

## Model selection

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
def maXr2_score(mod,X,Y):
    maX_r_score=0
    for r in range(42,100):
        X_train, X_test, Y_train, Y_test = train_test_split(X, Y,random_state = r,test_size=0.20)
        mod.fit(X_train,Y_train)
        Y_pred = mod.predict(X_test)
        r2_scr=r2_score(Y_test,Y_pred)
        print("r2 score corresponding to ",r," is ",r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r=r
    print("maX r2 score corresponding to ",final_r," is ",max_r_score)
    return final_r
```

#### Cross Validation

```python
#Cross_val_score fucntion 

from sklearn.model_selection import cross_val_score
def model_evaluation(model,X,Y):
    c_scores=cross_val_score(model,X,Y,cv=5,scoring="r2")
    print("Mean r2 score for regressor: ",c_scores.mean())
    print("standard deviation in r2 score for regressor: ",c_scores.std())
    print(c_scores)
```
